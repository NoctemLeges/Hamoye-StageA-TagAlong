```python
#import required modules and load the CSV file
import numpy as np
import pandas as pd
df = pd.read_csv("FoodBalanceSheets_E_Africa_NOFLAG.csv",encoding='latin-1')
df.info() #Print some basic info
```

```python
#Q.3 Total Protein Supply in Madagascar in 2015
#We filter for Area=Madagascar and Element=Protein Supply Quantity. Then we take out the Series corresponding to 2015 and sum over the values
df[(df['Area']=='Madagascar') & (df['Element']=='Protein supply quantity (g/capita/day)')]['Y2015'].sum()
```

```python
#Q.4 What is the mean and standard deviation across the whole dataset for the year 2017 to 2 decimal places?
print("Mean: ",df['Y2017'].mean())
print("Standard Deviation: ",df['Y2017'].std())
```

```python
#Q.6 Select columns ‘Y2017’ and ‘Area’, Perform a groupby operation on ‘Area’.  Which of these Areas had the 7th lowest sum in 2017?
grouped = df[['Area','Y2017']].groupby('Area').sum() #Grouped the selected columns based on Area and invoked sum(),which summed Y2017 column.
grouped.sort_values(by='Y2017') #Sorted the resulting dataframe based on values in Y2017
```

```python
#Q.11 Select columns ‘Y2017’ and ‘Area’, Perform a groupby operation on ‘Area’.  Which of these Areas had the highest sum in 2017?
grouped = df[['Area','Y2017']].groupby('Area').sum() #Selected the columns and grouped by Area,invoked sum which summed the Y2017 column.
grouped.sort_values(by='Y2017',ascending=False) # Sorted the resulting dataframe in Descending order.
```

```python
#Q.12 Perform a groupby operation on ‘Element’.  What year has the highest sum of Stock Variation?
grouped = df.groupby('Element').sum() #Group df by 'Element' column and take the sum of all columns in the resulting dataframe
grouped[grouped['Area Code']==571566][['Y2014','Y2015','Y2016','Y2017','Y2018']]# filter out the row corresponding to StockVariation and the columns for 2014,2015 and so on.
```

```python
#Q.17 What is the total number and percentage of missing data in 2014 to 3 decimal places?
numNA = df['Y2014'].isna().sum() #Find total number of missing data in 'Y2014'
percentNa = numNA/len(df['Y2014']) * 100 #Find percentage of missing data in 'Y2014'
print("Number of missing data: ",numNA)
print("Percentage of missing data: ",percentNa)
```
